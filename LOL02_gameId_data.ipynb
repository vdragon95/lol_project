```
import requests
import pandas as pd
import numpy as np
import time

api_key = 'RGAPI-5160c2f2-ceb9-4355-ac33-b2ebc1ae9aa6'

# queue = 'RANKED_SOLO_5x5'
# tier = 'PLATINUM' # PLATINUM, DIAMOND
# division = 'I' # I, II, III, IV
tier = 'grandmaster' # master, challenger
```

```
# <Match data> 조회를 위한 gameId얻기

# tier_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/' + tier + '/' + 
#                 tier + '_' + division + '_01tier_data.csv', encoding='cp949')
tier_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/' + tier + '/' + \
                tier + '_01tier_data.csv', encoding='cp949')


# Match Data 받아오기
# 10 시즌의 데이터만을 이용할 것이며 league_df => 기존에 수집한 account_id가 있는 league_df
match_info_df = pd.DataFrame()
season = str(13)

tier_df = tier_df.dropna(subset=['accountId']) # accountId가 float('na') 인 경우 행 삭제 (작업 참여 x)

print('request {} match data...'.format(len(tier_df)))
for i in range(len(tier_df)): # 한 account당 100판 가져옴
    try:
        match0 = 'https://kr.api.riotgames.com/lol/match/v4/matchlists/by-account/' + tier_df['accountId'].iloc[i]  +'?season=' + season + '&api_key=' + api_key
        r = requests.get(match0)
        
        while r.status_code == 429:
            time.sleep(5)
            match0 = 'https://kr.api.riotgames.com/lol/match/v4/matchlists/by-account/' + tier_df['accountId'].iloc[i]  +'?season=' + season + '&api_key=' + api_key
            r = requests.get(match0)
        
        matches = pd.DataFrame(r.json()['matches'])
        matches['summonerName'] = tier_df['summonerName'].iloc[i]

        # 칼럼 순서 변경 (summonerName 먼저오도록)
        cols = matches.columns.tolist()

        tmp = cols[-1] #11번째 column을
        del cols[-1]
        cols.insert(0,tmp) #2번째 위치로
        matches = matches[cols]
        print(matches)

        # drop_idx = matches[matches['queue'] != 420].index
        # matches = matches.drop(drop_idx)

        match_info_df = pd.concat([match_info_df, matches])
        print(i, 'load success')
    
    except Exception as e:
        print('[err.]:',i,'번째', e)
print('match_info_data File successfully loaded')
```
